```python
import os, getpass, pandas as pd
from sqlalchemy import create_engine, text

STUDENT_DB_URL = os.environ.get("NEON_STUDENT_DB_URL") or getpass.getpass("Paste STUDENT DB URL (hidden): ")
engine = create_engine(STUDENT_DB_URL, pool_pre_ping=True, future=True)
```

```python
with engine.begin() as conn:
    q1 = """
    SELECT p.maker, l.model, l.ram, l.price
    FROM laptop l
    JOIN product p ON l.model = p.model
    ORDER BY l.ram DESC, l.price DESC;
    """
    display(pd.read_sql(q1, conn))
```

```python
with engine.begin() as conn:
    q2 = """
    SELECT p.maker, pr.model, pr.type, pr.color, pr.price
    FROM printer pr
    JOIN product p ON pr.model = p.model
    WHERE pr.price < 500
    ORDER BY pr.price;
    """
    display(pd.read_sql(q2, conn))
```

```python
with engine.begin() as conn:
    q4 = """
    SELECT p.maker, l.model, l.screen, l.price
    FROM laptop l
    JOIN product p ON l.model = p.model
    WHERE l.screen >= 15 AND l.price < 1500
    ORDER BY l.price;
    """
    display(pd.read_sql(q4, conn))
```

```python
with engine.begin() as conn:
    q5 = """
    SELECT type, COUNT(*) AS num_models
    FROM product
    GROUP BY type
    ORDER BY num_models DESC;
    """
    display(pd.read_sql(q5, conn))
```
